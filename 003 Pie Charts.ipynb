```python
%config InlineBackend.figure_formats = ['svg']
from matplotlib import pyplot as plt
import numpy as np
```

```python
slices = [60,40, 15, 10]
labels = ['Amazon', 'Flipkart', 'Myntra', 'Others']
colors = ['#088fd5', '#fc4f30', '#e5ae37', '#6d904f']

plt.pie(slices,labels=labels, colors = colors)
plt.title('Pie Charts')
plt.show()
```

```python
plt.pie(slices, labels=labels,startangle= 90,
        wedgeprops = {'edgecolor':'#444444'} )
plt.title('Pie Charts')
plt.show()
```

```python
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
slices = [59219, 55466, 47544, 36443, 35917]
explode = [0,0,0,0.1,0]
plt.pie(slices, labels=labels, explode = explode, shadow = True,
        wedgeprops = {'edgecolor':'#000000'})
plt.title('Pie Charts')
plt.show()
```

```python
# to add percentage autopct= '%1.1f%%'
plt.pie(slices, labels=labels, explode = explode, shadow = True,
        autopct= '%1.1f%%',
        wedgeprops = {'edgecolor':'#000000'})

plt.title('Pie Charts')
plt.show()
```
