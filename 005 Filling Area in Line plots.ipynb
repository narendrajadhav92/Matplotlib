```python
%config InlineBackend.figure_formats = ['svg']
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data_s.csv')
ages = data['Age']
dev_y = data['All_Devs']
py_y = data['Python']
js_y = data['JavaScript']

plt.plot(ages, dev_y,color = '#444444' ,label = 'All Devs', linestyle= '--')
plt.plot(ages , py_y , 'b',label = 'Python')

plt.title('Median Salary(USD)')
plt.xlabel('Ages')
plt.ylabel('Salary')

plt.fill_between(ages, py_y,alpha = 0.25 )
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev_y,color = '#444444' ,label = 'All Devs', linestyle= '--')
plt.plot(ages , py_y , 'b',label = 'Python')

plt.title('Median Salary(USD)')
plt.xlabel('Ages')
plt.ylabel('Salary')

plt.fill_between(ages, py_y,median,alpha = 0.25,
                where = (py_y>median), interpolate = True)
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev_y,color = '#444444' ,label = 'All Devs', linestyle= '--')
plt.plot(ages , py_y , 'b',label = 'Python')

plt.title('Median Salary(USD)')
plt.xlabel('Ages')
plt.ylabel('Salary')
# 2 fill between
plt.fill_between(ages, py_y,median,alpha = 0.25,interpolate = True,
                where = (py_y>median))
plt.fill_between(ages, py_y,median,alpha = 0.25,color = 'red',
                where = (py_y<=median))
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev_y,color = '#444444' ,label = 'All Devs', linestyle= '--')
plt.plot(ages , py_y , 'b',label = 'Python')

plt.title('Median Salary(USD)')
plt.xlabel('Ages')
plt.ylabel('Salary')
# 2 fill between
plt.fill_between(ages, py_y,dev_y,alpha = 0.25,interpolate = True,
                where = (py_y>dev_y), label= 'Above Avg')
plt.fill_between(ages, py_y,dev_y,alpha = 0.25,color = 'red',interpolate = True,
                where = (py_y<=dev_y), label = 'Below Avg')
plt.legend()
plt.show()
```
