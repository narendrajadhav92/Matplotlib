```python
%config InlineBackend.figure_formats = ['svg']
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('seaborn')
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
# s is size argument, c= color
plt.scatter(x,y, s= 100, c= 'r',marker= 'o')
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```

```python
plt.scatter(x,y, s= 100, c= 'r',linewidth= 1, 
            edgecolor= 'k', alpha = 0.55)
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```

```python

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x,y, s= 100,c = colors,
            cmap= 'Greens')
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```

```python
plt.scatter(x,y, s= 100,c = colors,
            cmap= 'Greens')
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x,y, s=sizes,c = colors,
            cmap= 'Greens',edgecolor = 'k')
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```

```python
df = pd.read_csv('data_7.csv')
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']

plt.scatter(view_count,likes,
            edgecolor = 'k')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```

```python

plt.scatter(view_count,likes,
            edgecolor = 'k')

plt.xscale('log')
plt.yscale('log')
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```

```python
plt.scatter(view_count,likes,c= ratio ,cmap= 'summer',
            edgecolor = 'k')
cbar = plt.colorbar()
cbar.set_label('Like/Dislike ratio')
plt.xscale('log')
plt.yscale('log')
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```
